```python
def agrupador(Age):
    return "Adult" if Age <=40 else "Older Adult"

df["Age_group"]= df["Age"].apply(agrupador)
df
```

```python
print(df["Sleep Duration"].min(),df["Sleep Duration"].max())
```

```python
# Calidad y duracion del sueño de los generos
df.groupby("Gender")[["Sleep Duration","Quality of Sleep"]].mean()
```

Aca podemos ver que las mujeres tienen mayor duración de sueño y tambien calidad de sueño, segun nuestros datos

```python
# Nivel de estres, Calidad y duracion del sueño de las distintas ocupaciones 
df.groupby("Occupation")[["Stress Level","Sleep Duration", "Quality of Sleep"]].mean().sort_values(by="Quality of Sleep")
```

 Podemos observar las distintas profesiones y su promedio en la duración y calidad del sueño, esto tiene un relación con el nivel de estres en cada profesión 

```python
# agrupamos duración y calidad del sueño en base a la actividad fisica
df.groupby("Physical Activity Level")[["Sleep Duration","Quality of Sleep"]].mean()
```

En esta observación podemos concluir que no hay una relación considerable entre la actividad fisica y la calidad de sueño

```python
df["BMI Category"].value_counts()
```

```python
df.dtypes["BMI Category"]
```

```python
#convierto a la columna en typo string
df["BMI Category"] = df["BMI Category"].astype("string")
```

```python
#Nos aseguramos que los valores esten en minuscula y no haya espacios
df["BMI Category"] = df["BMI Category"].str.strip().str.lower()
```

```python
#Realizamos un map para asignar los valores correspondientes a la nueva columna
bmi_map={
    "normal": 1,
    "normal weight": 2,
    "overweight": 3,
    "obese": 4
}
df["BMI_numeric"]=df["BMI Category"].map(bmi_map)
df
```

```python
#Agrupamos por profesion y calculamos el promedio
df.groupby("Occupation")["BMI_numeric"].mean().sort_values()
```

Aca podes ver que la profesion con el mayor indice de masa corporal es "Sales Representative"
